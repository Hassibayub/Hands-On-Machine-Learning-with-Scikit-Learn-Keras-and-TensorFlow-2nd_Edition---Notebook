# Create test set

```python
def split_train_test(data, ratio, random_state=101, shuffle=True):
    
    np.random.seed(random_state)
    shuffle_indices = np.random.permutation(len(data))
    print("shuffle_indices:", shuffle_indices)
    
    size = int(len(data)*0.2)
    train_indices = shuffle_indices[size:]
    test_indices = shuffle_indices[:size]
    
    print("train_indices", len(train_indices))
    print("test_indices", len(test_indices))
    
    return data.iloc[train_indices], data.iloc[test_indices]

train, test = split_train_test(housing, 0.2, random_state=101, shuffle=True)
```

```python
def test_set_check(identifier, ratio = 0.2):
    return crc32(np.int64(identifier)) & 0xffffffff < ratio*2**32
    
def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
housing_with_id = housing.reset_index()
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'index')


len(train_set), len(test_set)
```

# Custom Transformer

```python
total_rooms_ix, total_bedrooms_ix, poplulation_ix, households_ix = 3, 4, 5, 6
class CombinedAttributeAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_rooms = True):
        self.add_bedrooms_per_rooms = add_bedrooms_per_rooms
    
    def fit(self, X, y=None):
        return self
    def fit_transform(self, X, y=None):
#         print("type: ", type(X))
        rooms_per_household= X[:,total_rooms_ix] / X[:,households_ix]
        population_per_household = X[:,poplulation_ix] / X[:,households_ix]
        
        if self.add_bedrooms_per_rooms:
            bedrooms_per_room = X[:,total_bedrooms_ix] / X[:,total_rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
```

```python
attr_adder = CombinedAttributeAdder(add_bedrooms_per_rooms=True)

housing_extra_attribs = attr_adder.fit_transform(housing.values)
```
